```python
# 

import pandas as pd
import numpy as np
import time
from datetime import date
import os



def find_earlier_eNodebList():
    list1=['','','']
    path= './data_eNodebList'
    date_list = os.listdir(path)
    n=0
    now1= date.today() 
    for i in range(60):   
        date_name_4G = 'eNodebList_4G({}).xlsx'.format(now1)
        date_name_5G = 'eNodebList_5G({}).xlsx'.format(now1)
        date_name_3G = 'eNodebList_3G({}).xlsx'.format(now1)
        if date_name_4G in date_list:
            if list1[0] =='':
                list1[0] = date_name_4G
                n=n+1
        if date_name_5G in date_list:
            if list1[1] =='':
                list1[1] = date_name_5G
                n=n+1
        if date_name_3G in date_list:
            if list1[2] =='':
                list1[2] = date_name_3G
                n=n+1
        if n==3:
            break
        else:
            now1 = now1 - datetime.timedelta(days=1)
    return(list1)

#-----------------------寫入檔案-------------------------------------#
today = date.today() 
writer =pd.ExcelWriter('./results(latlng)/bts_latlng_check({}).xlsx'.format(today)) 
#-------------------1. 讀取 3G 站台資料 & 分析 -----------------------#  
file_3G = find_earlier_eNodebList()[2] 
file1_path = "./data_eNodebList/{}".format(file_3G)
file_3G_df = pd.read_excel(file1_path,dtype=str)
file_3G_df = file_3G_df[['基地台編號(baseid)','細胞名稱(Cell_Name)','細胞經度(Cell_Lon)','細胞緯度(Cell_Lat)']]

###---------尋找 3G 本身資料錯誤----------------------###
selftest_3G_df = file_3G_df.drop_duplicates(subset=None,keep='first')
self_3G_df = selftest_3G_df.copy()
self_3G_df['3G_name'] =  self_3G_df['基地台編號(baseid)']+'('+self_3G_df['細胞名稱(Cell_Name)']+')'
self_3G_df = self_3G_df[self_3G_df.duplicated(subset="3G_name",keep=False)]
self_3G_df = self_3G_df.sort_values(by=['基地台編號(baseid)'])
self_3G_df.drop(columns='3G_name',inplace = True)

self_3G_df.to_excel(writer,sheet_name ='self_test_3G',index = False)
worksheet = writer.sheets['self_test_3G']
worksheet.set_column("A:D",21)

###-------------------------------------------------###

file_3G_df.drop_duplicates(subset=["基地台編號(baseid)","細胞名稱(Cell_Name)"],keep='first',inplace=True)
file_3G_df.reset_index(inplace=True, drop=True)
file_3G_df['3G_name'] = file_3G_df['基地台編號(baseid)']+'('+file_3G_df['細胞名稱(Cell_Name)']+')'
file_3G_df.rename(columns={'細胞名稱(Cell_Name)':'id_no'}, inplace=True)
file_3G_df['id_no'] = file_3G_df['id_no'].str.strip()
file_3G_df = file_3G_df.drop(['基地台編號(baseid)'],axis=1)
file_3G_df = file_3G_df[['id_no','3G_name','細胞經度(Cell_Lon)','細胞緯度(Cell_Lat)']]
file_3G_df.rename(columns={'細胞經度(Cell_Lon)': '3G_Cell_lng','細胞緯度(Cell_Lat)':'3G_Cell_lat'}, inplace=True)

# file_3G_df.to_excel(writer,sheet_name ='3G',index = False)

#-------------------2. 讀取 4G 站台資料 & 分析 -----------------------#  
file_4G = find_earlier_eNodebList()[0] 
file1_path = "./data_eNodebList/{}".format(file_4G)
file_4G_df = pd.read_excel(file1_path,dtype=str)
file_4G_df = file_4G_df[['基地台編號(baseid)','細胞名稱(Cell_Name)','細胞東經(Cell_Lon)','細胞北緯(Cell_Lat)']]

##---------尋找 4G 本身資料錯誤----------------------##
selftest_4G_df = file_4G_df.drop_duplicates(subset=None,keep='first')
self_4G_df = selftest_4G_df.copy()
self_4G_df['4G_name'] =  self_4G_df['基地台編號(baseid)']+'('+self_4G_df['細胞名稱(Cell_Name)']+')'
self_4G_df = self_4G_df[self_4G_df.duplicated(subset="4G_name",keep=False)]
self_4G_df = self_4G_df.sort_values(by=['基地台編號(baseid)'])
self_4G_df.drop(columns='4G_name',inplace = True)

self_4G_df.to_excel(writer,sheet_name ='self_test_4G',index = False)
worksheet = writer.sheets['self_test_4G']
worksheet.set_column("A:D",21)
##-------------------------------------------------##

file_4G_df.drop_duplicates(subset=["基地台編號(baseid)","細胞名稱(Cell_Name)"],keep='first',inplace=True)
file_4G_df.reset_index(inplace=True, drop=True)
file_4G_df['4G_name'] = file_4G_df['基地台編號(baseid)']+'('+file_4G_df['細胞名稱(Cell_Name)']+')'
file_4G_df.rename(columns={'細胞名稱(Cell_Name)': 'id_no'}, inplace=True)
file_4G_df['id_no'] = file_4G_df['id_no'].str.strip()
file_4G_df = file_4G_df.drop(['基地台編號(baseid)'],axis=1)                          
file_4G_df = file_4G_df[['id_no','4G_name','細胞東經(Cell_Lon)','細胞北緯(Cell_Lat)']]
file_4G_df.rename(columns={'細胞東經(Cell_Lon)': '4G_Cell_lng','細胞北緯(Cell_Lat)':'4G_Cell_lat'}, inplace=True)

# file_4G_df.to_excel(writer,sheet_name ='4G',index = False)


#-----------------3. 讀取 5G 站台資料 & 分析-------------------------#  

file_5G = find_earlier_eNodebList()[1]
file1_path = "./data_eNodebList/{}".format(file_5G)
file_5G_df = pd.read_excel(file1_path,dtype=str)
file_5G_df = file_5G_df[['基地台編號(BaseID)','細胞名稱(Cell_Name)','細胞東經(Cell_Lon)','細胞北緯(Cell_Lat)']]

###---------尋找 5G 本身資料錯誤----------------------###
selftest_5G_df = file_5G_df.drop_duplicates(subset=None,keep='first')
self_5G_df = selftest_5G_df.copy()
self_5G_df['5G_name'] =  self_5G_df['基地台編號(BaseID)']+'('+self_5G_df['細胞名稱(Cell_Name)']+')'
self_5G_df = self_5G_df[self_5G_df.duplicated(subset="5G_name",keep=False)]
self_5G_df = self_5G_df.sort_values(by=['基地台編號(BaseID)'])
self_5G_df.drop(columns='5G_name',inplace = True)

self_5G_df.to_excel(writer,sheet_name ='self_test_5G',index = False)
worksheet = writer.sheets['self_test_5G']
worksheet.set_column("A:D",21)
###-------------------------------------------------###

file_5G_df.drop_duplicates(subset=["基地台編號(BaseID)","細胞名稱(Cell_Name)"],keep='first',inplace=True)
file_5G_df.reset_index(inplace=True, drop=True)
file_5G_df['5G_name'] = file_5G_df['基地台編號(BaseID)']+'('+file_5G_df['細胞名稱(Cell_Name)']+')'                         
file_5G_df.rename(columns={'細胞名稱(Cell_Name)': 'id_no'}, inplace=True)
file_5G_df['id_no'] = file_5G_df['id_no'].str.strip()
file_5G_df = file_5G_df.drop(['基地台編號(BaseID)'],axis=1)                         
file_5G_df = file_5G_df[['id_no','5G_name','細胞東經(Cell_Lon)','細胞北緯(Cell_Lat)']]
file_5G_df.rename(columns={'細胞東經(Cell_Lon)': '5G_Cell_lng','細胞北緯(Cell_Lat)':'5G_Cell_lat'}, inplace=True)

# file_5G_df.to_excel(writer,sheet_name ='5G',index = False)
#----------------- 比對3/4/5G 經緯度---------------#

full_df = pd.merge(file_3G_df,file_4G_df,on='id_no',how='outer')
full_df = pd.merge(full_df,file_5G_df,on='id_no',how='outer')
full_df = full_df.fillna(value='')

for_3G_lng = full_df['3G_Cell_lng'].tolist()
for_3G_lat = full_df['3G_Cell_lat'].tolist()
for_4G_lng = full_df['4G_Cell_lng'].tolist()
for_4G_lat = full_df['4G_Cell_lat'].tolist()
for_5G_lng = full_df['5G_Cell_lng'].tolist()
for_5G_lat = full_df['5G_Cell_lat'].tolist()
    
lng_result =[]
for i in range(len(full_df['id_no'])):
    init_value=[for_3G_lng[i],for_4G_lng[i],for_5G_lng[i]]
    temp =list(filter(None, init_value))
    temp1=set(temp)
    if len(temp1) == 1:
        lng_result.append('O')
    else:
        lng_result.append('X') 

lat_result =[]
for i in range(len(full_df['id_no'])):
    init_value=[for_3G_lat[i],for_4G_lat[i],for_5G_lat[i]]
    temp =list(filter(None, init_value))
    temp1=set(temp)
    if len(temp1) == 1:
        lat_result.append('O')
    else:
        lat_result.append('X')        
        
        
full_df = full_df.assign(lng_check = lng_result)    
full_df = full_df.assign(lat_check = lat_result)  
full_df = full_df.sort_values(by=['id_no'])

full_df.to_excel(writer,sheet_name ='inter_check',index = False)
worksheet = writer.sheets['inter_check']
worksheet.set_column("A:B",19)
worksheet.set_column("C:D",13)
worksheet.set_column("E:E",19)
worksheet.set_column("F:G",13)
worksheet.set_column("H:H",19)
worksheet.set_column("I:J",13)
worksheet.set_column("K:L",10)
writer.save()
```
